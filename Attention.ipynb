```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sentence
sentence = "dream and big and work for it"
words = sentence.split()

# Create random 7 x 3 embedding
np.random.seed(2)
embeddings = np.random.rand(7, 3)

# 3D Visualization
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

x = embeddings[:, 0]
y = embeddings[:, 1]
z = embeddings[:, 2]

ax.scatter(x, y, z)

# Add word labels
for i, word in enumerate(words):
    ax.text(x[i], y[i], z[i], word)

ax.set_xlabel("Dimension 1")
ax.set_ylabel("Dimension 2")
ax.set_zlabel("Dimension 3")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sentence
sentence = "dream and big and work for it"
words = sentence.split()

# Create random 7 x 3 embedding (tensor)
np.random.seed(2)
embeddings = np.random.rand(7, 3)

# Compute magnitude (L2 norm) for each word tensor
magnitudes = np.linalg.norm(embeddings, axis=1)

# Print magnitudes
print("Word-wise Tensor Magnitudes:\n")
for word, mag in zip(words, magnitudes):
    print(f"{word:10s}  {mag:.4f}")

# 3D Visualization
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

x = embeddings[:, 0]
y = embeddings[:, 1]
z = embeddings[:, 2]

ax.scatter(x, y, z)

# Add word labels
for i, word in enumerate(words):
    ax.text(x[i], y[i], z[i], word)

ax.set_xlabel("Dimension 1")
ax.set_ylabel("Dimension 2")
ax.set_zlabel("Dimension 3")

plt.show()
```
