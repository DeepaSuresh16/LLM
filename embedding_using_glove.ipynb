```python
!pip install gensim numpy
import numpy as np
from gensim.models import KeyedVectors
import os

# Download and unzip GloVe file if not already present
if not os.path.exists("glove.6B.100d.txt"):
    print("Downloading GloVe embeddings...")
    # -nc (no-clobber) to prevent overwriting existing files
    !wget -nc http://nlp.stanford.edu/data/glove.6B.zip
    # -o (overwrite existing files without prompt)
    !unzip -o glove.6B.zip
    print("GloVe embeddings downloaded and unzipped.")

# Path to downloaded GloVe file
glove_file = "glove.6B.100d.txt"

# Load GloVe embeddings directly using KeyedVectors.load_word2vec_format
# with no_header=True as suggested by the DeprecationWarning
print(f"Loading GloVe embeddings from {glove_file}...")
glove_model = KeyedVectors.load_word2vec_format(glove_file, binary=False, no_header=True)
print("GloVe embeddings loaded.")

# Get embedding for a word
embedding = glove_model["king"]
print("Embedding for 'king':")
print("Embedding shape:", embedding.shape)
print(embedding)

# Example: find most similar words
print("\nWords most similar to 'king':")
print(glove_model.most_similar("king"))
```
